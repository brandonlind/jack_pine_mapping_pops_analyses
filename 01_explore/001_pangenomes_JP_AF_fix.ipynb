```python
lview, dview = get_client('default')
```

```python
# read in snptable
jp = pd.read_csv('/data/projects/pool_seq/pangenome/JP_pangenome/JP_pooled/snpsANDindels/02_baseline_filtered/JP_pooled-varscan_all_bedfiles_SNP.txt',
                sep='\t')
jp.head()
```

```python
# read in ploidy data
p = pd.read_csv('/data/projects/pool_seq/pangenome/JP_pangenome/SAMPLEFILE_headeridx.txt', sep='\t')
p.head()
```

```python
# create dict from ploidy data
ploidy = {}
for pop in p.columns:
    ploidy[pop] = p.loc[0, pop]
len([x for x in ploidy.values() if x != 40])
```

```python
# isolate freq columns
freqcols = [col for col in jp.columns if '.FREQ' in col]
len(freqcols)
```

```python
# chunks of rows
def get_af_chunks(args):
    df, rows = args
    afs = []
    for row in rows:
        # reduce to pops with no missing data
        freqs = df.loc[row, freqcols].str.replace("%", "").astype(float)/100
        freqs = freqs[~freqs.isnull()]
        # get pops
        pops = [freq.split(".")[0] for freq in freqs.keys()]
        # get counts of ALT across pops
        altcount = sum([freqs[pop]*ploidy[pop.split(".")[0]] for pop in freqs.keys()])
        # get total ploidy across pops
        tploidy = sum([v for pop,v in ploidy.items() if pop in pops])
        # get AF
        afs.append(altcount/tploidy)
       
    return rows, afs

dview['get_af_chunks'] = get_af_chunks
dview['ploidy'] = ploidy
dview['freqcols'] = freqcols
```

```python
# CHUNKS: calculate new allele freqs based on variable ploidy
jp['new_AF'] = ''
jobs = []
rows = []
for row in nb(jp.index):
    if sum(jp.loc[row, freqcols].isnull()) > 0:
        rows.append(row)
    if len(rows) == 10000:
        df = jp.loc[rows, :]
        jobs.append(lview.apply_async(get_af_chunks, (df, rows)))
        rows = []
```

```python
len(rows)
```

```python
# add last set of rows
df = jp.loc[rows, :]
jobs.append(lview.apply_async(get_af_chunks, (df, rows)))
```

```python
watch_async(jobs)
```

```python
for j in nb(jobs):
    if j.ready():
        rows, afs = j.r
        jp.loc[rows, 'new_AF'] = afs
```

```python
from scipy.stats import pearsonr
```

```python
pearsonr(jp.loc[jobs[0].r[0], 'AF'], jp.loc[jobs[0].r[0], 'new_AF'])
```

```python
# look at correlation
for j in jobs:
    rows = j.r[0]
    print(pearsonr(jp.loc[rows, 'AF'], jp.loc[rows, 'new_AF']))
```

```python
# replace AF data 
for j in nb(jobs):
    rows, afs = j.r
    jp.loc[rows, 'AF'] = afs
```
